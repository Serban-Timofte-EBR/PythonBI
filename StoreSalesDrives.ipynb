#Multilinear Regression Prep

```python
#Prepare X & Y
y = df.tsales
x = df.drop(columns = ["tsales"])
x = sm.add_constant(x)
x.head()
```

```python
#Training and Test Sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,
                                                    test_size=0.2,
                                                    random_state=1502)
```

##Multilinear Regression

```python
#Regression
model = sm.OLS(y_train, x_train).fit()
print(model.summary())
```

Only the size is statistically significant (margin is also close)

```python
#predictions
predictions = model.predict(x_test)
predictions[:4]
```

```python
#accuraccy assesment
from sklearn.metrics import mean_squared_error, mean_absolute_error
print("MAE: ", round(mean_absolute_error(y_test, predictions),1))
print("RMSE: ", round(np.sqrt(mean_squared_error(y_test, predictions)),1))
```

Our model is not really good because of the outliers
