```python
#Libraries
import pandas as pd
import seaborn as sns
```

```python
#load dataset
df = pd.read_csv("Baseball.csv")
df.head()
```

#Mean


```python
#Mean of Runs Scored (RS)
df.RS.mean()
```

```python
#Mean of Runs Scored (RS) - for ARI team
df.loc[df.Team == "ARI"].RS.mean()
```

```python
#Mean of Runs Scored (RS) by the Arizona team (ARI) since 2005
df.loc[(df.Team == "ARI") & (df.Year > 2005)].RS.mean()
```

```python
#Mean of Runs Allowed (RA) for Chicago (CHC) until 2007
df.loc[(df.Team == "CHC") & (df.Year<2007)].RA.mean()
```

#Median

```python
#Median and Mean of Wins (W)
print(df.W.mean())
df.W.median()
```

```python
#Median of Wins (W) of Baltimore Team (BAL) until 2000
df.loc[(df.Team == "BAL") & (df.Year <= 2000)].W.median()
```

#Mode

```python
#Mode, Median and Mean of OBP 
print(df.OBP.mode())
print(df.OBP.median())
print(df.OBP.mean())
```

```python
#Mode of OBP during 2010
print(df.loc[df.Year == 2010].OBP.mode())
```

#Correllation

```python
df.head()
```

```python
#pick variables
df_correllation = df[["RS", "RA", "W"]]
df_correllation.head(1)
```

```python
#Correlation Matrix
df_correllation.corr()
```

```python
#Correlation Heatmap
sns.heatmap(data = df_correllation.corr(),
            annot = True,
            fmt = '.2g',
            center = 0,
            cmap = 'coolwarm',
            linewidth = 1,
            linecolor = 'black')
```

```python
#Correlation Matrix between OBP, SLG and BA + heatmap with 2 parameters changed
df_new_correllation = df[["OBP", "SLG", "BA"]]
df_new_correllation.head(1)
```

```python
df_new_correllation.corr()
```

```python
sns.heatmap(data = df_new_correllation.corr(),
            center = 0.7,
            cmap = 'Greens',
            annot = True,
            fmt = '.1g',
            linewidth = 2,
            )
```
