```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
```

```python
df = pd.read_csv("CreditCard.csv")
df.head()
```

#Data Preparation


```python
#Tranform char
df = pd.get_dummies(df, drop_first = True)
df.head(2)
```

#Gaussian Mixture Model

```python
#Finding optimal nr. of clusters ( AIC / BIC )
```

```python
#Prepare
n_components = np.arange(1, 10)
n_components

#Create GMM model
models = [GaussianMixture(n_components = n,
                          random_state = 1502).fit(df) for n in n_components]

#Plot
plt.plot(n_components,
         [m.bic(df) for m in models],
         label = 'BIC')

plt.plot(n_components,
         [i.aic(df) for i in models],
         label = 'AIC')

plt.legend()
plt.xlabel('Nr. of Components')
```

```python
#Choose between 3 and 8 ( min values ).
```

```python
#Gaussian Mixture Model
model = GaussianMixture(n_components = 3,
                        random_state=1502).fit(df)
```

#Interpretation

```python
#Predict the cluster for each customer
cluster = pd.Series(model.predict(df))
cluster[:2]
```

```python
#Create cluster variable in df
df['Cluster'] = cluster
df.head(3)
```

```python
#Check how many people per cluster
df['Cluster'].value_counts()
```

```python
pd.DataFrame(model.means_,
             columns = list(df.iloc[:, :-1].columns))
```
